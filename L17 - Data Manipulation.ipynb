```python
# Long vs Wide data format
# Wide format is where we have a single row for every data point with multiple columns to hold the values of various attributes.

# Long format is where, for each data point we have as many rows as the number of attributes and each row contains the value of a particular attribute for a given data point.
```

```python
# melt --> wide to long
# Pivot --> long to wide
pd.DataFrame({"ece" : [102] , "cse" : [203] , "meac" : [300]}).melt(var_name= "branch" , value_name="number of students")
```

```python
pd.DataFrame(
    {
        "branch" : ["ece" , "cse" , "mech"],
        "2020" : ["123" , "230" , "140"],
        "2021" : ["234" , "345" , "374"]
    }
).melt(id_vars=["branch"] , var_name= "batch" , value_name="students")
```

```python
# Merge tha data
students = pd.read_excel("Dataset/joins/students.xlsx")
courses = pd.read_excel("Dataset/joins/courses.xlsx")
month2 = pd.read_excel("Dataset/joins/reg-month2.xlsx")
```

```python
# inner joins
students.merge(month2 , how = "inner" , on = "student_id")
```

```python
# left joins
courses.merge(month2 , how= "left" , on = "course_id")
```

```python
# right joins
month2.merge(courses , how = "right" , on = "course_id")
```

```python
courses
```

```python
# Full outer Join
courses.merge(month2 , how = "outer" , on = "course_id")
```

```python
# total revnew generated 
month2.merge(courses , how = "inner" , on = "course_id")["price"].sum()
```

```python
month2.merge(courses , how = "inner" , on = "course_id").groupby("course_id")["price"].sum().plot(kind= "bar")
```
